````python
class Cube:
    '''Class to calculate the cube given a number'''
    def __init__(self, value):
        self.x = value
    def calc_cube(self):
        cube = (self.x ** 3)
        return cube
    
````

````python
test = Cube(3)
print(test.x)
print(test.calc_cube())
````

## More about constructor method - init()
It's a special method that runs automatically when a class is instantied.<br>
Syntax:
`def __init__(self, [parameters]):
    #code
`<br>
self works like a reference for the class

````python
class Car:
    def __init__(self, brand, name, model, year, is_broken=None):
        self.brand = brand
        self.name = name
        self.model = model
        self.year = year
        self.is_broken = is_broken #It's none by default, which prevent errors in the case of status being forgotten to be passed

    def _car_status(self):
        if (self.is_broken is None):
            return "has status unknown"
        elif (not self.is_broken):
            return "is ready to road"
        else:
            return "needs to go to mechanic"
        
    def show_information(self):
        print(f"Your car is an {self.brand} {self.name}, model {self.model}, year {self.year}")
        print("Your car " + self._car_status())
````

````python
my_car = Car("Chevrolet", "Onix", "LT 1.0", 2014, False)

my_car.show_information()

my_car._car_status() #this works but is not recommended
````

## Attributes 
**Class Variable:** belongs to the class itself<br>
**Instance Variable:** belongs to each object individually

````python
class Motobike:
    vehicle_type = "2 Wheels" #Class variable

    def __init__(self, brand, name, cc):
        self.brand = brand
        self.name = name
        self.cc = cc
        #they are instance variables 
    
    def show_info(self):
        print(f"Motobike: {self.brand} {self.name} {self.cc}")
    
````

````python
my_bike = Motobike("Honda", "Hornet", "600")
new_bike = Motobike("Yamaha", "Tenere", "700")

my_bike.show_info()
print(f"{my_bike.vehicle_type}\n")
new_bike.show_info()
print(new_bike.vehicle_type)
````
