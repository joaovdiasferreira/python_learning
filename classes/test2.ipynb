# Testing Objects

To use class's methods and attributes, we need to derive an object of this class -> instantiation<br>
Syntax:<br>
```python
objectName = ClassName([parameters])
```
Using previous example:

````python
class Motobike:
    vehicle_type = "2 Wheels" #Class variable

    def __init__(self, brand, name, cc):
        self.brand = brand
        self.name = name
        self.cc = cc
        #they are instance variables 
    
    def show_info(self):
        print(f"Motobike: {self.brand} {self.name} {self.cc}")
````

````python
Motobike.show_info()
Motobike.vehicle_type
#It's not possible to use methods and attributes without an object
````

````python
#Now let's creat an object
my_bike = Motobike("Honda","Titan","160") #here I created an object called my_bike
print(my_bike.vehicle_type)
my_bike.show_info()

print(type(my_bike))
````

# Methods getter and setter

````python
class Test:
    def __init__(self, value):
        self.value = value

    def get_value(self):
        return self.value

    def set_value(self, new_value):
        self.value = new_value
````

````python
test = Test(7)

print(test.get_value())

test.set_value(10)

print(test.get_value())
````
